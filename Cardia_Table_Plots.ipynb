```python
# entire_heart_diease
total = 3783
num_ones = 79
entire_heart_diease = [1] * num_ones + [0] * (total - num_ones)

## Restaurant
exposed = 3041
never = 742
exposed_num_ones = 50
never_num_ones = 29
restaurant_exposure = ['Exposed'] * exposed_num_ones + ['Never'] * (exposed - exposed_num_ones) + ['Exposed'] * never_num_ones + ['Never'] * (never - never_num_ones)

## Bar
exposed = 2543
never = 1240
exposed_num_ones = 36
never_num_ones = 43
bar_exposure = ['Exposed'] * exposed_num_ones + ['Never'] * (exposed - exposed_num_ones) + ['Exposed'] * never_num_ones + ['Never'] * (never - never_num_ones)

## Workplace
exposed = 2454
never = 1329
exposed_num_ones = 35
never_num_ones = 44
workplace_exposure = ['Exposed'] * exposed_num_ones + ['Never'] * (exposed - exposed_num_ones) + ['Exposed'] * never_num_ones + ['Never'] * (never - never_num_ones)


data = [entire_heart_diease, restaurant_exposure, bar_exposure, workplace_exposure]

import pandas as pd
df = pd.DataFrame(data)
df = df.T
df.columns = ['Heart_Disease', 'Restaurant_exposure', 'Bar_exposure', 'Workplace_exposure']
df = df.sample(frac=1).reset_index(drop=True)

df.head(30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for exposure_type in ['Restaurant_exposure', 'Bar_exposure', 'Workplace_exposure']:
    sns.countplot(x='Heart_Disease', data=df[df[exposure_type]=='Exposed'])

    # Set labels and title
    plt.xlabel('Heart Disease')
    plt.ylabel('Count')
    plt.title(f'Distribution of Heart Disease among {exposure_type}')

    # Show the plot
    plt.show()
```
